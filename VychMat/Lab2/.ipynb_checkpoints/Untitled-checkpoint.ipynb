```python
import math as m

eps = 0.001
d = 0.01
n = 0
x0 = 50
c = 1

z = x = x0

while (abs(z - x) > eps or abs(m.exp(-x) - x + 2) > d):
    y = x + c * (m.exp(-x) - x + 2)
    z = x
    n += 1
    #print(f"{abs(m.exp(-y) - y + 2):.4f}")
    print(f"{n} {x:.4f} {y:.4f} {abs(y - x):.4f} {abs(m.exp(-y) - y + 2):.4f}")
    x = y
```

```python
import math as m
def func(x):
    return m.exp(-x) - x + 2

def expx(x):
    return m.exp(-x)

def linear(x):
    return x - 2

print("Общая")
for x in [x / 10 for x in range(40)]:
    print(x, func(x))
    
print("exp(-x)")
for x in [x / 10 for x in range(40)]:
    print(x, expx(x))
    
print("x-2")
for x in [x / 10 for x in range(40)]:
    print(x, linear(x))
```
